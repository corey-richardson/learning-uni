# Classify the Iris Data

First, import the required modules.

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.neighbors import KNeighborsClassifier
```

Load the Iris dataset.

```python
data = load_iris()
inputs = data.data
targets = data.target
```

We need to reduce the dimensionality of the data using PCA to visualise it. 

```python
pca = PCA(n_components=2)
compressed = pca.fit_transform(inputs)
```

Next, train the classifier itself. Instantiate the KNearestNeighbors classifier (set it for 10 neighbours). Then fit
the input data to the target values and use the predict method to see what the model prediction is for each of
the 150 flowers.

```python
classifier = KNeighborsClassifier(n_neighbors=10)
classifier.fit(inputs, targets)
classified_data = classifier.predict(inputs)
```

Finally, having obtained predictions for each flower, visualise the results. Produce a 2D scatter plot and colour
the points according to its predicted value.

```python
plt.figure()
plt.scatter(compressed[:,0], compressed[:,1], c=classified_data)
plt.title("PCA Projection")
plt.savefig("iris_pca.png", bbox_inches="tight")
plt.show()
```

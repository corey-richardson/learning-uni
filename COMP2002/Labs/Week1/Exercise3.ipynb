# More Numpy and Matplotlib

Generate 100 random numbers from the uniform distribution (`np.random.rand`) and 100 samples from the
normal distribution (`np.random.randn`). Produce a boxplot showing the two distributions.

```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
uniform_distribution = np.random.rand(100)
normal_distribution = np.random.randn(100)
```

```python
plt.boxplot([uniform_distribution, normal_distribution])
plt.xticks([1,2], ["Uniform", "Normal"])
plt.show()
```

Produce a histogram showing the two distributions.

```python
plt.hist(uniform_distribution)
plt.xlabel("Uniform")
plt.show()

plt.hist(normal_distribution)
plt.xlabel("Normal")
plt.show()
```
